## Q. Whether there is any relationship between sepal length and sepal.width of flowers?

```python
plt.scatter(df['sepal.length'],df['sepal.width'])
plt.show()
```

### From scatter plot above it is difficut to identify whether there is any relationship between sepal length and sepal width of flower. Hence, we will try to see if particular variety of flowers have any relationship between them.

```python
sns.scatterplot(x="sepal.length", y="sepal.width", hue="variety", data=df)
plt.xlabel('sepal.length',fontsize=30, color='r')
plt.ylabel('sepal.width', fontsize=30, color='r')
plt.show()
```

## on scatter plot above we can say that variety named setosa has lesser sepal length but much better sepal width than other two varieties.
## Verginica has highest sepal length amonst three varieties
